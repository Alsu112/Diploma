```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img2 = Image.open('.PNG')
img2.convert('RGB').save('result.png')
img = Image.open('result.png')

width = img.size[0]
height = img.size[1]
print(width)
print(height)

width_m = 1000
height_m = 0.8

width_coef= width_m/width
height_coef= height_m/height

print(img.format, img.size, img.mode) #    

pixels = list(img.getdata())
#print(pixels)
result = []
for pix in pixels:
    r, g, b = pix
    result.append(r + g + b)
#print(result)
index = []
for number in range(len(result)):
    if result[number] == 0:
        index.append(number)
amount_black = len(index)
coordinate = []
coordinate_x = []
coordinate_y = []
print(index)
for number in range(len(index)):
    m = index[number] // width #     
    l = index[number] - m * width #    
    m = height - index[number] // width
    coordinate.append((m * height_coef,l * width_coef))
    coordinate_x.append(l * width_coef)
    coordinate_y.append(m * height_coef)
#print(coordinate)
#print(coordinate_x)
#print(coordinate_y)
f = open('text2.txt', 'w')
for items in coordinate_y:
    f.write('%f\n' %items)
f.close()
f2 = open('text.txt', 'w')
for items in coordinate_x:
    f2.write('%f\n' %items)
f2.close()

#print(coordinate_x)
#print(coordinate_y)
    
    
#print(result)
#print(len(result))
#my_result = np.zeros((img.size[1], img.size[0]))
#for m in range(img.size[1] - 1):
#    for l in range(img.size[0] - 1):
#        my_result[m][l] = result[(m + 1)*(l+1)]
#print(my_result)
    
```
